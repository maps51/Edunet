```python
df=df.drop('education', axis=1)
b=pd.get_dummies(df.workclass).add_prefix('workclass_')
c=pd.get_dummies(df['marital-status']).add_prefix('marital-status_')
d=pd.get_dummies(df['native-country']).add_prefix('native-country_')
e=pd.get_dummies(df.relationship).add_prefix('relationship_')
f=pd.get_dummies(df.race).add_prefix('race_')
```

```python
df=pd.concat([df.drop('workclass', axis=1), b], axis=1)
df=pd.concat([df.drop('marital-status', axis=1), c], axis=1)
df=pd.concat([df.drop('native-country', axis=1), d], axis=1)
df=pd.concat([df.drop('relationship', axis=1), e], axis=1)
df=pd.concat([df.drop('race', axis=1), f], axis=1)
```

```python
df
```

```python
df['gender']=df['gender'].apply(lambda x: 1 if x=='Male' else 0)
df['income']=df['income'].apply(lambda x: 1 if x=='>50K' else 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df=df.drop('fnlwgt', axis=1)
```

```python
train_df, test_df=train_test_split(df, test_size=0.2)
```

```python
train_df
```

```python
test_df
```

```python
train_X=train_df.drop('income', axis=1)
train_y=train_df['income']
test_X=test_df.drop('income', axis=1)
test_y=test_df['income']
```

```python
forest=RandomForestClassifier()
forest.fit(train_X, train_y)
```

```python
g=forest.score(test_X, test_y)
print("Accuracy : ", g)
```

```python
forest.feature_importances_
```

```python
importances=dict(zip(forest.feature_names_in_, forest.feature_importances_))
importances={k: v for k, v in sorted(importances.items(), key=lambda x:x[1], reverse=True)}
```

```python
importances
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    'n_estimators':[50, 100, 250],
    'max_depth' :[5, 10, 30, None],
    'min_samples_split': [2, 4],
    'max_features': ['sqrt', 'log2']
}

grid_search=GridSearchCV(estimator=RandomForestClassifier(), param_grid=param_grid, verbose=10)
```

```python
grid_search.fit(train_X, train_y)
```

```python
grid_search.best_estimator_
```

```python
forest=grid_search.best_estimator_
```

```python
h=forest.score(test_X, test_y)
print("Accuracy : ", h)
```

```python
importances=dict(zip(forest.feature_names_in_, forest.feature_importances_))
importances={k: v for k, v in sorted(importances.items(), key=lambda x:x[1], reverse=True)}
```
